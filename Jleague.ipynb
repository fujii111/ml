```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X = np.array([3, 6, 1, 1, 1, 4, 6, 11, 5, 3, 5, 1, 2])
T = np.array([15433, 16239, 19714, 21617, 20966, 16156, 15381, 16419, 17665, 16423, 19103, 20467, 19434])
X_min = 0
X_max = 12

# データグラフ表示の関数
def show_data(X, T):
    plt.figure(figsize=(8, 4))
    plt.plot(X, T, linestyle="none", marker="o", markeredgecolor="k")
    plt.xlim(X_min, X_max)
    plt.grid(True)

show_data(X, T)
plt.show()

# 直線表示関数
def show_line(w):
    xb = np.linspace(X_min, X_max, 10)
    y = w[0] * xb + w[1]
    plt.plot(xb, y)
    
# 平均二乗誤差の計算を行う関数
def mse_line(x, t, w):
    y = w[0] * x + w[1]
    mse = np.mean((y - t) ** 2)
    return mse

show_data(X, T)
W=(-600, 21000)
show_line(W)
plt.show()

print(mse_line(X, T, W))

# 直線モデルのパラメータを解析的に求める関数
def fit_line(x, t):
    mx = np.mean(x)
    mt = np.mean(t)
    mtx = np.mean(t * x)
    mxx = np.mean(x * x)
    w0 = (mtx - mt * mx) / (mxx - mx ** 2)
    w1 = mt - w0 * mx
    return np.array([w0, w1])

show_data(X, T)
W = fit_line(X, T)
show_line(W)
plt.show()

print(mse_line(X, T, W))

print(W[0] * 3 + W[1])
```
